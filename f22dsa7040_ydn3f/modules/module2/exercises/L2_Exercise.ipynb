**Activity 3:** Use ellipse library to create an **elliptical correlation plot** for the diamonds data using carat, price, table, and depth variables. Make sure to use a **colorblind safe, diverging** color scheme. 

```r
library(ellipse)
head(diamonds)
data=diamonds[,c("carat","depth","price","table")]
## Your answer for activity 3 goes here ---

# (first pick and display the color palette, then display the plot)
my_pal <- colorRampPalette(brewer.pal(9, 'BuPu'))(50)

#correlation matrix
cor_data <- cor(data)

#ordered indeces
ord <- order(cor_data[1, ])
ord_cor_data = cor_data[ord, ord]
plotcorr(ord_cor_data, col=my_pal[ord_cor_data*25+25],mar=c(1,1,1,1))
```

**Activity 4:** Create a scatter plot using **all** of the diamonds data set for carat vs. price and encode `clarity` variable with a **sequential** color scheme. 

```r
## Your answer for activity 4 goes here ---

ggplot(diamonds, aes(carat, price, color=clarity)) + geom_point() + scale_colour_brewer(type='seq', palette=4) + theme_dark()
```

**Activity 5:** Repeat Activity 4 with an **alpha transparency** added to the plot. **Is this a better idea? Discuss your answer.**

```r
## Your answer for activity 5 goes here ---


ggplot(diamonds, aes(carat, price, color=clarity)) + geom_point(alpha=0.5) + scale_colour_brewer(type='seq', palette=4) + theme_dark()

#Adding alpha (transparency) is often recommended because it makes clusters or groupings easier to see
#as opposed to no transparency which can cover up some of the colors (data) behind superimposed layers of colors.
```
